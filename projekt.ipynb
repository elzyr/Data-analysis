## b) Wizualizacje danych

````python
numeric_features = data.select_dtypes(include=["number"])
corr_matrix = numeric_features.corr()


mask = np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
corr_pairs = (
    corr_matrix.where(mask)
    .stack()
    .reset_index()
    .rename(
        columns={"level_0": "Variable 1", "level_1": "Variable 2", 0: "Correlation"}
    )
)

top_10 = corr_pairs.nlargest(10, "Correlation")
bottom_10 = corr_pairs.nsmallest(10, "Correlation")

results = (
    pd.concat([bottom_10, top_10])
    .sort_values("Correlation", ascending=True)
    .reset_index(drop=True)
)

norm = plt.Normalize(results["Correlation"].min(), results["Correlation"].max())
colors = plt.cm.coolwarm(norm(results["Correlation"]))

plt.figure(figsize=(12, 8))
pairs = results["Variable 1"] + " - " + results["Variable 2"]
plt.barh(pairs, results["Correlation"], color=colors)
plt.xlabel("Korelacja")
plt.title("10 Najniższych i 10 Najwyższych Korelacji")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
````

Wykres pokazuje, że wyniki z pierwszego semestru („**Curricular units 1st sem (grade)**”) są silnie skorelowane z wynikami w drugim semestrze („**Curricular units 2nd sem (grade)**”), co sugeruje, że studenci utrzymują swój poziom nauki.

Można również zauważyć, że starsze osoby częściej wybierają tryb wieczorowy („**Daytime/evening attendance - Age at enrollment**”), co może wynikać z ich zobowiązań zawodowych lub rodzinnych.

Dodatkowo, osoby zadłużone („**Debtor - Tuition fees up to date**”) mogą mieć trudności z terminowym opłacaniem czesnego, co potencjalnie wpływa na ich dalszą edukację.

````python
tak = data["Target"].value_counts()


def format_labels(pct, all_values):
    absolute = int(round(pct / 100.0 * sum(all_values)))
    return f"{pct:.1f}%\n({absolute})"


plt.figure(figsize=(8, 8))
plt.pie(
    tak, labels=tak.index, autopct=lambda pct: format_labels(pct, tak), startangle=90
)
plt.title("Procentowy udział wartości w Target (z liczbami)")
plt.show()
````

Największą grupę stanowią absolwenci („Graduate”), których liczba wynosi 2209. Znacząca część studentów, 1421 osób, nie ukończyła studiów i została sklasyfikowana jako rezygnujący („Dropout”). Najmniej liczną grupą są osoby nadal kontynuujące naukę („Enrolled”), których liczba wynosi 794, co może sugerować, że większość studentów albo ukończyła edukację, albo z niej zrezygnowała.

W tym kroku usunąłem ewentualne spacje z nazw kolumn, aby uniknąć problemów z późniejszym przetwarzaniem danych. Następnie zmieniłem wartości w kolumnie „Target” z  tekstowych na wartości liczbowe: „Dropout” na 0, „Enrolled” na 1, a „Graduate” na 2, co umożliwia ich późniejszą analizę 

````python
data.columns = data.columns.str.strip()
data["Target"] = (
    data["Target"].replace({"Dropout": 0, "Enrolled": 1, "Graduate": 2}).astype(int)
)
````

Następnym etapem jest zamiana zmiennych kategorycznych na format one-hot encoding, co oznacza, że każda unikalna wartość z tych kolumn została przekształcona w osobną binarną zmienną. Dzięki temu dane są lepiej przygotowane do analizy i trenowania modeli.

````python
categorical_features = [
    "Marital status",
    "Application mode",
    "Course",
    "Daytime/evening attendance",
    "Previous qualification",
    "Nacionality",
    "Mother's qualification",
    "Father's qualification",
    "Mother's occupation",
    "Father's occupation",
    "Application order",
]
data = pd.get_dummies(data, columns=categorical_features, prefix=categorical_features)
````

## b) Wybór cech

Aby wybrać najważniejsze cechy, zastosuję dwa podejścia: **korelację Pearsona** oraz test **Chi-kwadrat**. Następnie porównam uzyskane wyniki i wyodrębnię **część wspólną**, aby uzyskać zestaw najbardziej istotnych cech do dalszej analizy.

````python
X = data.drop(columns=["Target"])
y = data["Target"]

number_of_features = 30
````

````python
scaler_std = StandardScaler()
X_scaled = scaler_std.fit_transform(X)
X_df_scaled = pd.DataFrame(X_scaled, columns=X.columns)

pearson_corr = X_df_scaled.corrwith(y).abs()
pearson_selected = pearson_corr.sort_values(ascending=False).head(number_of_features)
print("\nNajważniejsze cechy wg korelacji Pearsona:")
pearson_selected.index.tolist()
````

````python
X["Inflation rate"] = X["Inflation rate"].clip(lower=0)
X["GDP"] = X["GDP"].clip(lower=0)

chi_selector = SelectKBest(score_func=chi2, k=number_of_features)
chi_selector.fit(X, y)
chi_scores = pd.Series(chi_selector.scores_, index=X.columns)
chi_selected = chi_scores.sort_values(ascending=False).head(number_of_features)

print("\nNajważniejsze cechy wg testu Chi-kwadrat:")
chi_selected.index.tolist()
````

````python
pearson_set = set(pearson_selected.index)
chi_set = set(chi_selected.index)
common_features = list(pearson_set & chi_set)

X_selected = X[common_features]
print("\nCzęść wspólna:")
common_features
````

## c) Przekształcenia danych (np. normalizacja, dyskretyzacja)

Po wybraniu najistotniejszych cech dokonałem ich normalizacji za pomocą ```MinMaxScaler```, który przekształca wartości na zakres od 0 do 1. Taki zabieg pozwala zapewnić jednolitą skalę danych, co może poprawić wynik.

````python
scaler_minmax = MinMaxScaler()
X_selected_norm = scaler_minmax.fit_transform(X_selected)
X_selected_norm = pd.DataFrame(X_selected_norm, columns=common_features)
X_selected_norm.head()
````

# 4. Tworzenie modelu i ocena (algorytmów)

## a) Podziel zbiór danych (trening i test)

````python
X_train, X_test, y_train, y_test = train_test_split(
    X_selected_norm, y, test_size=0.3, random_state=0
)
````

## Las losowy

Najważniejsze parametry dla Random Forest Classifier to 
- ```n_estimators```, określający liczbę drzew w lesie 
- ```max_depth```, kontrolujący maksymalną głębokość drzew, co pomaga zapobiegać przeuczeniu 
- ```class_weight``` pozwalający na dostosowanie wag klas w przypadku niezrównoważonych danych

````python
class_labels = ["Dropout", "Enrolled", "Graduate"]

rf = RandomForestClassifier(
    n_estimators=50, max_depth=10, class_weight="balanced", random_state=0
).fit(X_train, y_train)
rf_pred = rf.predict(X_test)

print(f"Dokładność Random Forest: {accuracy_score(y_test, rf_pred) * 100:.2f}%")

sns.heatmap(
    confusion_matrix(y_test, rf_pred),
    annot=True,
    cmap="Blues",
    fmt="d",
    xticklabels=class_labels,
    yticklabels=class_labels,
)
plt.title("Las losowy")
plt.xlabel("Predykcja")
plt.ylabel("Rzeczywista klasa")
plt.show()
````

## SVC

Dla SVC kluczowe parametry to ```kernel```, definiujący funkcję jądra (np. 'linear', 'rbf', 'poly'), oraz ```C```, regulujący kompromis między poprawnym klasyfikowaniem danych a prostotą modelu (większa wartość C zmniejsza liczbę błędnych klasyfikacji, ale może prowadzić do przeuczenia).

````python
svc_model = SVC(kernel="linear", C=3.0, random_state=0).fit(X_train, y_train)
svc_pred = svc_model.predict(X_test)

print(f"Dokładność SVM: {accuracy_score(y_test, svc_pred) * 100:.2f}%")

conf_matrix = confusion_matrix(y_test, svc_pred)

sns.heatmap(
    conf_matrix,
    annot=True,
    fmt="d",
    cmap="Blues",
    linewidths=1,
    xticklabels=class_labels,
    yticklabels=class_labels,
)
plt.title("SVM")
plt.xlabel("Predykcja")
plt.ylabel("Rzeczywista klasa")
plt.show()
````

## Przedstaw wyniki oraz porównaj algorytmy według wskazanych metod oceny.

**Random Forest** osiągnął wyższą dokładność (**77.41%**) niż **SVM** (**76.20%**) przy wybranych parametrach.  

Model **lasu losowego** lepiej klasyfikuje studentów, którzy **zrezygnowali ze studiów**, osiągając większą liczbę poprawnych klasyfikacji tej klasy w porównaniu do SVM.  

**Las losowy** lepiej rozpoznał aktualnych studentów niż **SVM**, jednak żaden z modeli nie jest w stanie dobrze ich rozpoznać.

Oba modele dobrze sklasyfikowały **absolwentów**, jednak **SVM zrobił to bardziej precyzyjnie**, co może wynikać z większej liczby danych w tej kategorii.


## Wskaż lub wykonaj czynności poprawiające wynik.

Spróbuję poprawić wynik usuwając ```class_weight```, aby poprawić stopień w którym las poprawnie klasyfikuje absolwentów

````python
rf = RandomForestClassifier(n_estimators=100, max_depth=35, random_state=0).fit(
    X_train, y_train
)
rf_pred = rf.predict(X_test)

print(
    f"Dokładność Random Forest po poprawie: {accuracy_score(y_test, rf_pred) * 100:.2f}%"
)

sns.heatmap(
    confusion_matrix(y_test, rf_pred),
    annot=True,
    cmap="Blues",
    fmt="d",
    xticklabels=class_labels,
    yticklabels=class_labels,
)
plt.title("Las losowy")
plt.xlabel("Predykcja")
plt.ylabel("Prawdziwa klasa")
plt.show()
````

Zwiększenie liczby drzew w modelu (```n_estimators```) oraz regulacja głębokości (```max_depth```) pozwoliły na poprawę ogólnej dokładności modelu.  

Zmiana wartości ```class_weight``` wpłynęła na lepszą klasyfikację absolwentów, jednak pogorszyła predykcje dla pozostałych klas.
